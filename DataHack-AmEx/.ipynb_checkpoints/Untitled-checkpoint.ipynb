```python
import csv

count_of_rows = 0
files = os.listdir('../DataHack-AmEx/train')
print(files)

for f in files:
    with open('../DataHack-AmEx/train/' + f) as file:
        print()
        print("Fetching :" , f)
        csv_reader_object = csv.reader(file)
        for row in csv_reader_object:
            count_of_rows += 1
        print("File     : {0} has {1} rows".format(f,count_of_rows) )
        
```

```python
campaign_df_1 = pd.read_csv('../DataHack-AmEx/train/campaign_data.csv', delimiter=',' , nrows=1)
coupon_df_1 = pd.read_csv('../DataHack-AmEx/train/coupon_item_mapping.csv', delimiter=',' , nrows=1)
customer_dem_df_1 = pd.read_csv('../DataHack-AmEx/train/customer_demographics.csv', delimiter=',' , nrows=1)
customer_tran_df_1 = pd.read_csv('../DataHack-AmEx/train/customer_transaction_data.csv', delimiter=',' , nrows=1)                             
item_df_1 = pd.read_csv('../DataHack-AmEx/train/item_data.csv', delimiter=',' , nrows=1)
train_df_1 = pd.read_csv('../DataHack-AmEx/train/train.csv', delimiter=',' , nrows=1)  
```

```python
types_dict1 = {'campaign_id': 'int32' }

types_dict2 = {'coupon_id': 'int32',
             'item_id': 'int32' }

types_dict3 = {'customer_id': 'int32',
             'family_size': 'O',
             'income_bracket': 'int32',
             'no_of_children': 'O',
             'rented': 'int32' }

types_dict4 = {'coupon_discount': 'float32',
             'customer_id': 'int32',
             'item_id': 'int32',
             'other_discount': 'float32',
             'quantity': 'int16',
             'selling_price': 'float32' }

types_dict5 = {'brand': 'int32',
             'item_id': 'int32'}

types_dict6 = {'campaign_id': 'int32',
             'coupon_id': 'int32',
             'customer_id': 'int32',
             'id': 'int32',
             'redemption_status': 'int8'}
```

```python
campaign_df = pd.read_csv('../DataHack-AmEx/train/campaign_data.csv', delimiter=',' , low_memory=True, dtype = types_dict1 )
coupon_df = pd.read_csv('../DataHack-AmEx/train/coupon_item_mapping.csv', delimiter=',' , low_memory=True, dtype = types_dict2 )
customer_dem_df = pd.read_csv('../DataHack-AmEx/train/customer_demographics.csv', delimiter=',', low_memory=True, dtype = types_dict3 )
customer_tran_df = pd.read_csv('../DataHack-AmEx/train/customer_transaction_data.csv', delimiter=',' , low_memory=True, dtype = types_dict4 )                             
item_df = pd.read_csv('../DataHack-AmEx/train/item_data.csv', delimiter=',' , low_memory=True, dtype = types_dict5 ) 
train_df = pd.read_csv('../DataHack-AmEx/train/train.csv', delimiter=',' , low_memory=True, dtype = types_dict6 )                            
```

## Merging/ Joining data

```python
#merging train and campaign

train_camp_df = pd.merge(train_df 
                   , campaign_df
                   , left_on = 'campaign_id'
                   , right_on  = 'campaign_id'
                   , how = 'left' 
                  )

train_camp_df.head()
```

```python
train_camp_df.info()
```

```python
del(campaign_df)
```

```python
#merging customer_transaction and item_data

customer_df = pd.merge(customer_tran_df 
                   , item_df
                   , left_on = 'item_id'
                   , right_on  = 'item_id'
                   , how = 'left' 
                  )

customer_df.head()
```

```python
customer_df.info()
```

```python
customer_df.drop(['date'] , axis=1 , inplace = True)
customer_df.info()
```

```python
del(customer_tran_df)
del(item_df)
```

```python
#merging train_camp_df(created above) and customer_demographics

train_cust1 = pd.merge(train_camp_df 
                   , customer_dem_df
                   , left_on = 'customer_id'
                   , right_on  = 'customer_id'
                   , how = 'left' 
                  )

train_cust1.head()
```

```python
train_cust1.info()
```

```python
del(customer_dem_df)
```
